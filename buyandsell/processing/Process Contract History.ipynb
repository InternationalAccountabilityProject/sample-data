### Just for fun, make the table of contracts that don't overlap

```python
not_overlap = merged_reordered[merged_reordered.contract_number != merged_reordered['\xef\xbb\xbfcontract-number']]
print len(not_overlap)
#not_overlap['contract_number_prefix'] = not_overlap['\xef\xbb\xbfcontract-number'].str.split('-').str.get(0)
columns = list(not_overlap.columns)
#columns.insert(1, 'contract_number_prefix')
#columns.pop()
not_overlap = not_overlap[columns]
not_overlap.head()
```

```python
print len(not_overlap)
#print sum(not_overlap.contract_number_prefix.value_counts().head(150).values) / len(not_overlap)
percents = {}
for column in not_overlap.columns:
    percents[column] = sum(not_overlap[column].value_counts().head(150).values) / len(not_overlap)

sorted_percents = sorted(percents.iteritems(), key=operator.itemgetter(1))
sorted_percents
```

```python
print len(overlap)
#print sum(overlap.contract_number_prefix.value_counts().head(150).values) / len(overlap)
percents_o = {}
for column in overlap.columns:
    percents_o[column] = sum(overlap[column].value_counts().head(150).values) / len(overlap)

sorted_percents_o = sorted(percents_o.iteritems(), key=operator.itemgetter(1))
sorted_percents_o
```

```python
ratio = {}
for item in percents:
    ratio[item] = percents[item] / percents_o[item] * 100

sorted_ratio = sorted(ratio.iteritems(), key=operator.itemgetter(1))
sorted_ratio
```
