```python
awarded = pd.read_csv("../original/tpsgc-pwgsc_aa-a - awarded.csv")
```

```python
awarded[awarded['reference_number'] == 'PW-$$BB-241-23477']
```

```python
awarded = awarded[awarded['language'] == "English"]
```

```python
awarded
```

```python
refined = awarded[['contract_number','reference_number']]
```

```python
refined
```

```python
merged = pd.merge(english_data, refined, how='outer', left_on='\xef\xbb\xbfcontract-number', right_on='contract_number')
```

```python
merged_cols = list(merged.columns)
print merged_cols
merged_cols.insert(0, merged_cols[-2])
merged_cols.insert(0, merged_cols[-1])
merged_cols.pop()
merged_cols.pop()
print merged_cols
```

```python
merged_reordered = merged[merged_cols]
```

```python
merged_reordered.head()
```

```python
overlap = merged_reordered[merged_reordered.contract_number == merged['\xef\xbb\xbfcontract-number']]
```

```python
print len(overlap)
print len(overlap.contract_number.value_counts())
overlap.head()
```

```python
merged_to_awards = pd.merge(awarded, english_data, how='inner', right_on='\xef\xbb\xbfcontract-number', left_on='contract_number')
print len(merged_to_awards.contract_value.value_counts())
print len(awarded.contract_value.value_counts())
print awarded.columns
print merged_to_awards.columns
```

### Summarize and export to csv

```python
print "Number of rows in contract history = " + str(len(merged_reordered))
print "Number of rows where contract number is available in awards table for linking = " + str(len(overlap))
print "Percentage of contract history rows that can be linked = " + str(len(overlap) / len(merged_reordered) * 100) + "%"
print "Percentage of unique contract numbers that can be linked = " + str(len(overlap.contract_number.value_counts()) / len(merged_reordered.contract_number.value_counts()) * 100) + "%"
print "Percentage of awards with contract data = " + str(len(merged_to_awards.contract_value.value_counts()) / len(awarded.contract_value.value_counts()) * 100) + "%"
```
